```python
from lightkurve import search_targetpixelfile #why is this even needed? 

# find star from the kepler catalogue! 
pixelfile = search_targetpixelfile("KIC 8462852", cadence="long", quarter=16).download();
pixelfile.plot(frame=1);

```

```python
# turn it into a light curve! this is a graph that shows us 
import matplotlib.pyplot as plt
lc = pixelfile.to_lightcurve(aperture_mask='all');

lc.flux
lc.time
lc.plot()
plt.show()
```

Let's take a look at some of the data types we're going to be seeing - in more detail, of course. 

First up: The *Periodogram*! When analysing the spectra of objects, we're going to go off certain signals - though I don't really understand the algorithm that went behind it. In general, spikes in a periodograpm translate to a high *signal strength* - meaning that there's a lot of it in one place. We can find exoplanets by comparing two periodograms with each other - one of just the star and another with the exoplanet. This is how JWST finds the spectra of planets.

The second on the list is the *lightcurve* object - simply put, there's a benchmark 'expected' luminosity for the star to be at, and the light from the star is given as a ratio to that benchmark luminosity.

```python
# Next tutorial - using Kepler Files! - learning to import from kepler

from lightkurve import search_lightcurve
lc = search_lightcurve("KIC 8120608", author="Kepler", quarter=10, cadence="long").download().remove_nans()
lc.plot();
plt.show();
```

Let's understand some of the vocab we use here!
