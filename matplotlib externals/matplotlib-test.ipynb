```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# simple plotting - we can define and plot datasets! 

x= [1,2,3]
y= [1,2,3]

plt.plot(x,y)
plt.show() #don't forget this! 
```

```python
#add axis titles, decoration etc 
x= [1,2,3]
y= [1,2,3]

plt.plot(x,y, color="black")

plt.xlabel("bananas per apple")
plt.ylabel("apples per oranges")
plt.title("total fruits per fruit")

plt.show()
```

# We can make a grid contain much more than one plot! Use the .subplots() function. 

```python
# subplots, layout

x = [1,2,3,4] #make sure each dataset is corroborated!
y = [4,7,6,2]
z = [19,23,12.5,15]

fig, axs = plt.subplots(1,2) # fig is the entire diagram, axs denote the two graphs

fig, (ax1, ax2) = plt.subplots(1,2) # different way of defining the subplots

#set a total tile
fig.suptitle("Baseball Games per Ferrets hunted")

# background color total graph
fig.set_facecolor("aliceblue")

ax1.plot(x,y)
ax2.plot(y,z)
```

```python
#remember to give your axes proper names! 
#time for scatter plots. we're making two! 

baseballs = [1,2,3,4]
ferrets = [27,83,111,49]
trains = [72,234,112,9]

baseballpferrets, (baseferrets, ferretstrains) = plt.subplots(1,2)

#scatter plot with .scatter

baseferrets.scatter(baseballs, ferrets)
ferretstrains.scatter(ferrets, trains)

baseballpferrets.suptitle("Ferrets Playing Baseball with Train Cars at any given moment")

```
